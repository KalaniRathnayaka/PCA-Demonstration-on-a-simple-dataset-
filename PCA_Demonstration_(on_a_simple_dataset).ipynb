```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Load the Iris dataset
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target
```

```python
# Step 3: Show dataset information
print("Dataset Info:")
print(f"Total samples: {X.shape[0]}")
print(f"Total features (before PCA): {X.shape[1]}")
print(f"Number of species: {len(set(y))}")
print(f"Species labels: {set(y)} (0: setosa, 1: versicolor, 2: virginica)")
```

```python
# Standardize the data (important before PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Create a DataFrame with PCA results
df_pca = pd.DataFrame(data=X_pca, columns=["PC1", "PC2"])
df_pca['target'] = y
```

```python
# Plot the 2D PCA result
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_pca, x="PC1", y="PC2", hue="target", palette='Set2')
plt.title("PCA of Iris Dataset (4D â†’ 2D)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()
```
